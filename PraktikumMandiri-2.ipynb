```python
outlier_data = {
    'Gender': ["Male","Female","Female","Male"],
    'Item Purchased': ["Sweater","Sandals","Sweater","Sandals"],
    'Category': ["Clothing","Footwear","Clothing","Footwear"],
    'Purchase Amount (USD)': [60, 150, 72, 154],
    'Location': ["Jakrata", "Depok", "Bekasi", "Bogor"],
    'Subscription Status': ["Yes","Yes","Yes","Yes"],
    'Payment Method': ["Cash","Cash","Cash","Cash"],
    'Shipping Type': ["Express", "Express", "Express", "Express"],
    'Previous Purchases': [20, 85, 18, 89],
    'Frequency of Purchases': ["Weekly", "Weekly", "Weekly", "Weekly"]
}

outlier_df = pd.DataFrame(outlier_data)

df = pd.concat([df, outlier_df], ignore_index=True)
df
```

```python
df.boxplot(column=['Purchase Amount (USD)','Previous Purchases'])
plt.title('Boxplot of Purchase Amount (USD) and Previous Purchases with Outliers')
plt.show()
```

```python
# Hitung kuartil pertama (Q1)
q1 = df['Purchase Amount (USD)'].quantile(0.25)
print("Q1 : ", q1)
# Hitung kuartil ketiga (Q3)
q3 = df['Purchase Amount (USD)'].quantile(0.75)
print("Q3 : ", q3)
iqr = q3 - q1
# Hitung batas bawah (Lower Bound) dan batas atas (upper Bound) untuk Outlier
lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr

# temukan outlier dalam DaraFrame
outliers = df[(df['Purchase Amount (USD)'] < lower_bound) | (df['Purchase Amount (USD)'] > upper_bound)]

print("Outlier Purchase Amount (USD): ")
print(outliers)
# mencari data outlier dalam kolom
```
